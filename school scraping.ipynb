```python
from selenium import webdriver
import time
import json

BASE_URL = "https://www.greatschools.org/illinois/tinley-park/schools/?distance=60&lat=41.5596&lon=-87.835&st%5B%5D=public&zip=60487"
PAGE_URL = "https://www.greatschools.org/illinois/tinley-park/schools/?distance=60&lat=41.5596&lon=-87.835&page={}&st%5B%5D=public&zip=60487"
PAUSE_INTERVAL = 10  
SLEEP_TIME = 5  
FILENAME = 'progress.json'

def save_progress(text_list):
    with open(FILENAME, 'w') as f:
        json.dump(text_list, f)

def load_progress():
    try:
        with open(FILENAME, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        return []

text_list = load_progress()

# Set up the browser
options = webdriver.ChromeOptions()
options.headless = True  # This runs Chrome in the background (headless mode)
browser = webdriver.Chrome(options=options)

def get_text_by_color(browser, color_code):
    script = """
    let texts = [];
    let elements = [...document.querySelectorAll("*")].filter(el => window.getComputedStyle(el, null).color === arguments[0]);
    for (let el of elements) {
        texts.push(el.innerText);
    }
    return texts;
    """
    return browser.execute_script(script, color_code)


# Fetch and parse the unique first link if starting from scratch
if not text_list:
    browser.get(BASE_URL)
    texts = get_text_by_color(browser, "rgb(26, 157, 209)")  # CSS color codes are often in rgb format
    text_list.extend(texts)

start_page = 2 + len(text_list) // PAUSE_INTERVAL

for page in range(start_page, 100):
    browser.get(PAGE_URL.format(page))
    time.sleep(SLEEP_TIME)  # Allow the page to load fully
    texts = get_text_by_color(browser, "rgb(26, 157, 209)")
    text_list.extend(texts)
    print(f"Processed page {page}. Texts: {texts}")
    
    if page % PAUSE_INTERVAL == 0:
        print("Pausing... Press Enter to continue...")
        save_progress(text_list)
        input()

browser.quit()  
save_progress(text_list)  
print(text_list)
```

```python
words = text_list

# Using list comprehension to filter and modify the list
result = [word.lower() for word in words if 'school' in word.lower()]

print(result)
```
